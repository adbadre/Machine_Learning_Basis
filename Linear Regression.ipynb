```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Create fake data where y=5+2*x1+3*x2
x=np.random.normal([0,1], [0.1,0.5], (100,2))
y=np.array([2*x[:,0]+3*x[:,1]+5]).T
print(x.shape,y.shape)
```

```python
##Plot the data 
i=0
for col in x.T:
    fig,ax=plt.subplots(figsize = (5,5))
    ax.hist(col)
    i=i+1
```

```python
#Initialize the bias and the weight 
b=np.random.uniform(0,1,100)
w=np.random.uniform(0,0.5,2)

pred=[]
#Fit the algo ( No validation but who cares on fake data ^^)
for epoch in range(10):
    #Compute the prediction
    y_pred=np.add(np.dot(np.array([w]),x.T),b)
    pred.append(y_pred)
    #Compute the error
    mse=np.mean(((y_pred.T-y)**2)[:,0])
    print(mse)
    #Update the bias and weigth with the gradient
    w=w-0.1*2*(y_pred.T-y)[0]
    b=b-0.1*2*b
```

```python
#Plot the prediction and the target over the training
for i in range(10):
    fig,ax=plt.subplots(figsize = (5,5))
    ax.plot(y[:,0],label="Target")
    ax.plot(pred[i][0],label="Prediction")
    ax.set_xlabel("epoch")
    ax.set_ylabel("y")
    ax.legend(loc='upper right', shadow=True, fontsize='x-small')
    ax.set_title('Prediction and Target vs at epoch'+str(i))
```
